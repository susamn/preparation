# Kernel Density Estimation (KDE) Plots for Model Evaluation

Kernel Density Estimation (KDE) plots estimate the probability density function (PDF) of a dataset. They are extremely useful in regression analysis for visually comparing actual and predicted values. By providing a smooth approximation, they serve as a modern and effective replacement for Seaborn's deprecated `distplot`.

## Importing Libraries
Let's import the necessary modules for data manipulation, modeling, and plotting.

```python
import numpy as npy
import pandas as pds
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

## Generating Sample Data
We will create synthetic data with a linear relationship and some added Gaussian noise.

```python
# Generating Sample Data
npy.random.seed(42)
x = npy.random.rand(100) * 10
y = 3 * x + npy.random.normal(0, 3, 100)  # Linear relation with noise
data = pds.DataFrame({'X': x, 'Y': y})
```

## Splitting Data
Next, we prepare the dataset by splitting it into training (80%) and testing (20%) sets.

```python
# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(data[['X']], data['Y'], test_size=0.2, random_state=42)
```

## Training a Model
We train a simple `LinearRegression` model using the training set and then generate predictions for the test set.

```python
# Training a Model
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

## Plotting KDE for Observed vs. Predicted Values
Finally, we visualize the true distributions against the predicted ones. `kdeplot()` effectively smooths out the distributions so we can detect any deviations.

```python
# Plotting KDE for Observed vs. Predicted Values
plt.figure(figsize=(8, 5))
sns.kdeplot(y_test, label='Actual', fill=True, color='blue')
sns.kdeplot(y_pred, label='Predicted', fill=True, color='red')
plt.xlabel('Target Variable')
plt.ylabel('Density')
plt.title('KDE Plot of Actual vs. Predicted Values')
plt.legend()
plt.show()
```

## Interpretation of the KDE Plot
- **Overlap Between Distributions**: Significant overlap usually indicates the model successfully captured the general distribution of the variable.
- **Peak Differences (Mode Shifts)**: If the peaks don't horizontally align, the model may be misestimating ranges. If they don't align vertically, the concentration of predictions isn't quite right.
- **Spread of the Distributions**: A wider actual spread and narrower predicted spread implies the model is slightly underestimating variance (smoothing it too much).
- **Tails**: Alignment here tests if the model generates extreme outliers consistent with real-world observed data.
